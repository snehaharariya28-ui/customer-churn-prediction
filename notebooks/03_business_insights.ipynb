## Business Recommendations

Based on the analysis and predictive modeling results, the following recommendations can help reduce customer churn:

1. Review Pricing Strategy
   - Customers appear highly sensitive to energy prices and price fluctuations.
   - Consider competitive pricing plans and targeted discounts for high-risk customers.

2. Monitor High-Risk Customers
   - Use the churn prediction model to identify customers with a high probability of leaving.
   - Proactively engage these customers through retention campaigns.

3. Improve Customer Retention Programs
   - Offer loyalty rewards, personalized plans, and long-term contract benefits.
   - Focus retention efforts on valuable customers with high revenue contribution.

4. Track Price Variation Impact
   - Monitor how pricing changes affect customer behavior.
   - Evaluate customer response before implementing major pricing adjustments.

5. Develop Customer Segmentation Strategies
   - Group customers based on consumption patterns, profitability, and churn risk.
   - Create targeted retention strategies for each segment.

## Conclusion

This project analyzed customer churn behavior using exploratory data analysis, feature engineering, and machine learning techniques. Logistic Regression and Balanced Logistic Regression models were developed to predict customer churn.

The initial Logistic Regression model achieved high accuracy but failed to identify churn customers effectively due to class imbalance. To address this issue, a Balanced Logistic Regression model was implemented, significantly improving churn detection performance.

Feature importance analysis revealed that pricing-related variables, customer profitability metrics, and consumption patterns are key factors influencing churn. Based on these findings, several business recommendations were proposed, including pricing optimization, targeted retention strategies, and proactive monitoring of high-risk customers.

Overall, the project demonstrates how data-driven decision-making can help organizations identify churn risks and improve customer retention.

## Future Scope

The performance of the churn prediction system can be further improved by exploring advanced machine learning algorithms such as Random Forest, XGBoost, and Gradient Boosting. These models may capture complex relationships in customer behavior more effectively than Logistic Regression.

Additional customer information, such as customer support interactions, customer satisfaction scores, and demographic attributes, could also be incorporated to improve prediction accuracy. Furthermore, real-time churn monitoring systems can be developed to identify at-risk customers and enable proactive retention strategies.
